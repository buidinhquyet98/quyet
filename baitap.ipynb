```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from google.colab import drive 
sns.set_theme()
drive.mount('/content/gdrive',force_remount=True)
df=pd.read_csv('/content/gdrive/My Drive/colab/StudentsPerformance.csv')
df
```

```
df.dropna(axis=1,how='all',inplace=True)
df.dropna(axis=0,how='all',inplace=True)
df.head(10)
```

```
sns.relplot(data=df,x="reading score", y="writing score",hue="gender",style="gender")
```

```
sns.relplot(data=df,x="reading score", y="writing score",hue="gender",style="test preparation course")
```

```
sns.relplot(data=df,x="reading score", y="writing score",size="lunch",sizes=(15,50))
```

```
sns.relplot(data=df,x="reading score", y="writing score",hue="gender",col="test preparation course",row="gender")
```

```
sns.regplot(x="reading score", y="writing score",data=df)
```

```
sns.lmplot(x="reading score", y="writing score",data=df,hue="gender")
```
